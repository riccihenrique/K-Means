```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
import seaborn as sb

%matplotlib inline
```

```python
df = pd.read_csv('iris.csv') # Leitura do CSV
#df = pd.read_csv('heart.csv')
```

```python
sb.pairplot(df)
```

```python
#X = np.array(df.drop('variety', axis=1)) # Iris
X = np.array(df.drop('target', axis=1)) # Fruits
```

```python
X
```

```python
from sklearn.cluster import KMeans # Import do Kmeans
```

```python
#kmeans = KMeans(n_clusters=3, random_state=0) # Iris (3 Classes)
kmeans = KMeans(n_clusters=3, random_state=0) # Fruits (2 Classes)
```

```python
kmeans.fit(X) # Treino
```

```python
kmeans.labels_ # Labels
```

```python
def classfication(str1):
    if(str1 == 0):
        return "Versicolor" 
    else:
        if(str1 == 1):
            return "Setosa"
        else: 
            return "Virginica"

df['Kclasses'] = [classfication(x) for x in kmeans.labels_] # isso serve sÃ³ para as flores
#df['Kclasses'] = kmeans.labels_ # demais
```

```python
df
```

```python
c = 0;
for i, j in zip(df["target"], df["Kclasses"]):
    if i == j:
        c += 1
        
print(c / 150 * 100)  # Procentagem de acerto para as flores
```

```python
sb.pairplot(df, hue="Kclasses")
```
